## Exemplo

Calcule a média harmônica Amortecida dos seguintes
números: 3.6, 8.9, 10

A média harmônica é definida pela
equação:

$$H =\frac{N}{\sum_{i=1}^{N} \frac{1}{n_i+X}} - X $$


Onde $X = 4$

````
# Declaração de variaveis
n = [3.6,8.9,10]
X = 4
N = len(n) 
soma = 0

# Cálculo do média
for i in range(0,3):
  soma = soma + (1)/(n[i]+(X))
H = (N)/soma - X
print(H)
````
